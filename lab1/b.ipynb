```python
import numpy as np
import matplotlib.pyplot as plt
import subprocess
import os
```

```python
subprocess.run(['mk', '3.cpp'], check=True)
```

```python
def get_histogram_data(option):
    result = subprocess.run(['./build/3', option], 
                          capture_output=True, text=True, check=True)
    lines = result.stdout.strip().split('\n')
    
    data = []
    for line in lines:
        if line.startswith('#') or not line.strip():
            continue
        if ',' in line:
            value, prob = line.split(',')
            data.append((int(value), float(prob)))
    
    return data
```

```python
big_normal = get_histogram_data('--big-normal')
big_penalty = get_histogram_data('--big-penalty')
big_bonus = get_histogram_data('--big-bonus')
three_normal = get_histogram_data('--three-normal')
three_penalty = get_histogram_data('--three-penalty')
three_bonus = get_histogram_data('--three-bonus')
double_dice = get_histogram_data('--double-dice')
```

```python
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
values, probs = zip(*big_normal)
plt.bar(values, probs, alpha=0.7, color='blue')
plt.title('  [1,100]')
plt.xlabel('')
plt.ylabel('')
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 2)
values, probs = zip(*big_penalty)
plt.bar(values, probs, alpha=0.7, color='red')
plt.title('  [1,100]')
plt.xlabel('')
plt.ylabel('')
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 3)
values, probs = zip(*big_bonus)
plt.bar(values, probs, alpha=0.7, color='green')
plt.title('  [1,100]')
plt.xlabel('')
plt.ylabel('')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
